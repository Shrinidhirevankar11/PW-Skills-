## 1. What is a parameter?
- A parameter is a variable that defines a model’s internal configuration and is learned from the data during training.
For example, in a linear regression model y = mx + c, both m (slope) and c (intercept) are parameters.

##2. What is correlation?
- Correlation measures the statistical relationship between two variables, showing how one variable changes with respect to the other.

##3. What does negative correlation mean?

- A negative correlation means that as one variable increases, the other decreases.
Example: As the number of hours spent watching TV increases, exam scores may decrease.

##4. Define Machine Learning. What are the main components in Machine Learning?

- Machine Learning (ML) is a subset of Artificial Intelligence that enables systems to learn from data and make predictions or decisions without being explicitly programmed.
Main components:

- Data: Input for training.

- Model: Algorithm used for learning.

- Loss Function: Measures prediction error.

- Optimizer: Minimizes the loss.

- Evaluation: Measures model performance.

##5. How does loss value help in determining whether the model is good or not?

- A low loss value indicates that the model’s predictions are close to actual results, meaning the model performs well.
A high loss value means poor performance.

##6. What are continuous and categorical variables?

- Continuous Variables: Have numerical values within a range (e.g., height, salary).

- Categorical Variables: Represent categories or labels (e.g., gender, city, color).

##7. How do we handle categorical variables in Machine Learning? What are the common techniques?

- Common techniques:

- Label Encoding: Converts categories into numeric labels.

- One-Hot Encoding: Creates binary columns for each category.

- Ordinal Encoding: Assigns integer values to ordered categories.

##8. What do you mean by training and testing a dataset?

- Training Dataset: Used to train the model.

- Testing Dataset: Used to evaluate model performance on unseen data.

## 9. What is sklearn.preprocessing?

- sklearn.preprocessing is a module in Scikit-learn that provides utilities for scaling, transforming, and encoding features (like StandardScaler, LabelEncoder, etc.).

## 10. What is a Test set?

- A test set is a separate portion of data used to evaluate how well a trained model generalizes to new, unseen data.

##15. What is causation? Explain difference between correlation and causation with an example.

- Causation means one variable directly affects another.
Correlation just shows a relationship, not cause-effect.
Example: Ice cream sales and drowning deaths are correlated but not causally linked — both rise due to summer heat.

##16. What is an Optimizer? What are different types of optimizers? Explain each with an example.

An optimizer adjusts model parameters to minimize loss.
Common types:

- Gradient Descent: Basic optimization method.

- SGD (Stochastic Gradient Descent): Updates weights per sample.

- Adam: Combines momentum and adaptive learning rates.
- Example: optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

## 20. What is feature scaling? How does it help in Machine Learning?

- Feature scaling standardizes feature values to a similar range (e.g., 0–1 or mean 0 and variance 1).
It helps algorithms like gradient descent converge faster and prevents bias toward features with larger values.
